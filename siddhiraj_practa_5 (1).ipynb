now sorted the male data as well as female data to calculate outliers with the help of IQR method

```python
data= sorted(male_age_data)
data1=sorted(female_age_data)
data
```

Here write the code for Outlier using IQR method for male data and calculate upper bound , lower bound and print it

```python
import numpy as np
q1=np.percentile(data, 25)
q3=np.percentile(data, 75)
IQR = q3-q1
mub=q3+(1.5*IQR)
mlb=q1-(1.5*IQR)

```

```python
print(q1)
```

```python
print(mub)
```

```python
print(mlb)
```

Here write the code for Outlier using IQR method for female data and calculate upper bound , lower bound and print it

```python
q1=np.percentile(data1, 25)
q3=np.percentile(data1, 75)
IQR = q3-q1
ub=q3+(1.5*IQR)
lb=q1-(1.5*IQR)
```

```python
print(q1)
```

```python
print(ub)
```

```python
print(lb)
```

Now we found the outlier in male data 

```python
outlier = []
for i in male_age_data:
    if(i<mlb):
        outlier.append(i)
    if(i>mub):
        outlier.append(i)
print(outlier)
```

There is no any outlier in the female data so it print the null

```python
outlier = []
for i in female_age_data:
    if(i<lb):
        outlier.append(i)
    if(i>ub):
        outlier.append(i)
print(outlier)
```

A strip plot is a scatter plot where individual data points are plotted along a categorical axis. It helps visualize the distribution and density of data points while preserving each observation. here draw the stripplot which consist x axis as sex and y axis as age

```python
sns.stripplot(x='sex', y='age', data=df)
```

A violin plot combines a box plot and a density plot. here draw the violinplot which consist x axis as sex and y axis as age

```python
sns.violinplot(x='sex', y='age', data=df)
```

```python
sns.swarmplot(x='sex', y='age', data=df)
```
